## Exercício 1: Número Positivo

Crie um programa que recebe um número inteiro do usuário e imprime `"Positivo"` se o número for maior que zero, e `"Não Positivo"` caso contrário.

```python
numero = int(input("Digite um número inteiro: "))
if numero > 0:
    print("Positivo")
else:
    print("Não Positivo")
```

## Exercício 2: Número Par ou Ímpar

Escreva um programa que solicita um número inteiro e verifica se ele é par ou ímpar. Imprima `"Par"` ou `"Ímpar"` de acordo.

```python
numero = int(input("Digite um número inteiro: "))
if numero % 2 == 0:
    print("Par")
else:
    print("Ímpar")
```

## Exercício 3: Elegibilidade para Dirigir

Peça ao usuário sua idade e se ele possui carteira de motorista (responda com `True` ou `False`). Determine se a pessoa pode dirigir. Uma pessoa pode dirigir se tiver 18 anos ou mais E possuir carteira de motorista.

```python
idade = int(input("Digite sua idade: "))
possui_carteira = input("Você possui carteira de motorista? (responda com Sim ou Não): ").lower()
if idade >= 18 and possui_carteira == "sim":
    print("Você pode dirigir.")
else:
    print("Você não pode dirigir.")
```

## Exercício 4: Ano Bissexto

Receba um ano como entrada e determine se é um ano bissexto. Um ano é bissexto se for divisível por 4, exceto por anos que são divisíveis por 100, a menos que também sejam divisíveis por 400.

```python
ano = int(input("Digite um ano: "))
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print("O ano é bissexto.")
else:
    print("O ano não é bissexto.")
```

## Exercício 5: Elegibilidade para Desconto

Uma loja oferece um desconto especial. Um cliente é elegível se for estudante OU se tiver menos de 18 anos E estiver acompanhado por um adulto. Peça a idade do cliente, se ele é estudante (True/False) e se está acompanhado por um adulto (True/False). Imprima se o cliente é `"Elegível para Desconto"` ou `"Não Elegível"`.

```python
idade = int(input("Digite sua idade: "))
estudante = input("Você é estudante? (responda com Sim ou Não): ").lower()
esta_acompanhado = input("Você está acompanhado por um adulto? (responda com Sim ou Não): ").lower()
if estudante == "sim" or (idade < 18 and esta_acompanhado == "sim"):
    print("Elegível para Desconto")
else:
    print("Não Elegível")
```

## Exercício 6: Usuário

## Exercício 7: Validação de Reserva

Desenvolva uma função que verifica se uma reserva é válida. A função deve receber a data de início (string 'YYYY-MM-DD'), data de fim (string 'YYYY-MM-DD'), número de hóspedes (int) e se o tipo de quarto desejado está disponível (boolean). Uma reserva é válida se:
1. A data de início não é posterior à data de fim.
2. O número de hóspedes é maior que zero.
3. O tipo de quarto desejado está disponível.
Retorne `True` para reserva válida e `False` caso contrário. (Não se preocupe com validação de formato de data, apenas com a comparação lógica).

```python
inicio = input("Digite a data de início (YYYY-MM-DD): ")
fim = input("Digite a data de fim (YYYY-MM-DD): ")
hospedes = int(input("Digite o número de hóspedes: "))
disponivel = input("O tipo de quarto desejado está disponível? (responda com Sim ou Não): ").lower()
if inicio < fim and hospedes > 0 and disponivel == "sim":
    print("Reserva válida")
else:
    print("Reserva inválida")
```

## Exercício 8: Classificador de Condição Climática

Crie um programa que, dadas a temperatura (int), umidade (int) e velocidade do vento (int) em km/h, classifique a condição climática:
- Se a temperatura for maior que 25, a umidade menor que 60 E a velocidade do vento menor que 10, imprima `"Dia Ideal de Verão"`.
- Se a temperatura for menor que 0 OU houver previsão de neve (booleano), imprima `"Alerta de Inverno"`.
- Caso contrário, imprima `"Condição Normal"`.

```python
temperatura = int(input("Digite a temperatura: "))
umidade = int(input("Digite a umidade: "))
vento = int(input("Digite a velocidade do vento(km/h): "))
neve = input("Há previsão de neve? (responda com Sim ou Não): ").lower()
if temperatura > 25 and umidade < 60 and vento < 10:
    print("Dia Ideal de Verão")
elif temperatura < 0 or neve == "sim":
    print("Alerta de Inverno")
else:
    print("Condição Normal")
```

## Exercício 9: Regra de Elegibilidade Complexa

Implemente uma função que determina a elegibilidade para um programa social baseado nas seguintes regras:
Um indivíduo é elegível se:
- (Tiver 18 anos ou mais E renda mensal abaixo de 30000 E (estiver empregado OU for estudante))
OU
- (Tiver 65 anos ou mais E renda mensal abaixo de 50000).

A função deve receber `idade` (int), `renda_mensal` (int), `empregado` (bool), `estudante` (bool) e retornar `True` se for elegível, `False` caso contrário.

```python
idade = int(input("Digite sua idade: "))
renda_mensal = int(input("Digite sua renda mensal(reais): "))
empregado = input("Você está empregado? (responda com Sim ou Não): ").lower()
estudante = input("Você é estudante? (responda com Sim ou Não): ").lower()
if (idade >= 18 and renda_mensal < 30000 and (empregado == "sim" or estudante == "sim")) or (idade >= 65 and renda_mensal < 50000):
    print("Você é elegível")
else:
    print("Você não é elegível")
```
